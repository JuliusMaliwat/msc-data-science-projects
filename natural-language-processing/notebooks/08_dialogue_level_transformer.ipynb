## Setup

```python
import pandas as pd
import ast
import re
from transformers import RobertaTokenizer

df_train = pd.read_csv("train.csv")
df_val = pd.read_csv("validation.csv")
df_test = pd.read_csv("test.csv")

def fix_and_eval_list(s):
    if isinstance(s, list):
        return s
    s_fixed = re.sub(r"(\d)\s+(?=\d)", r"\1, ", s)
    return ast.literal_eval(s_fixed)

def parse_dialog(s):
    utterances = re.findall(r"'(.*?)'|\"(.*?)\"", s)
    return [u[0] if u[0] else u[1] for u in utterances]

# Applica parsing
for df in [df_train, df_val, df_test]:
    df["dialog"] = df["dialog"].apply(parse_dialog)
    df["act"] = df["act"].apply(fix_and_eval_list)
    df["emotion"] = df["emotion"].apply(fix_and_eval_list)
```

```python
def build_flat_df(df):
    rows = []
    for dialog_id, row in df.iterrows():
        for turn_id, (utt, emo) in enumerate(zip(row["dialog"], row["emotion"])):
            rows.append({
                "dialog_id": dialog_id,
                "turn_id": turn_id,
                "utterance": utt,
                "label": emo
            })
    return pd.DataFrame(rows)

df_train_flat = build_flat_df(df_train)
df_val_flat = build_flat_df(df_val)
df_test_flat = build_flat_df(df_test)

# Check shapes
print(df_train_flat.shape, df_val_flat.shape, df_test_flat.shape)
```

```python
import numpy as np
import pandas as pd

train_embeddings = np.load("train_embeddings.npy")
val_embeddings = np.load("val_embeddings.npy")
test_embeddings = np.load("test_embeddings.npy")

df_train_flat["embedding"] = list(train_embeddings)
df_val_flat["embedding"] = list(val_embeddings)
df_test_flat["embedding"] = list(test_embeddings)

def group_embeddings_by_dialog(df_flat):
    grouped_embeddings = []
    grouped_labels = []
    for dialog_id, group in df_flat.groupby("dialog_id"):
        group = group.sort_values("turn_id")
        try:
            emb = np.stack(group["embedding"].values)
        except:
            print(f"Errore nel dialog_id {dialog_id}, skippo")
            continue
        labels = group["label"].values
        if emb.shape[0] != len(labels):
            print(f"Dropping dialog_id {dialog_id} - embeddings: {emb.shape[0]}, labels: {len(labels)}")
            continue
        grouped_embeddings.append(emb)
        grouped_labels.append(labels)
    return grouped_embeddings, grouped_labels

train_dialog_embeddings, train_dialog_labels = group_embeddings_by_dialog(df_train_flat)
val_dialog_embeddings, val_dialog_labels = group_embeddings_by_dialog(df_val_flat)
test_dialog_embeddings, test_dialog_labels = group_embeddings_by_dialog(df_test_flat)

# Controllo
print(train_dialog_embeddings[0].shape, train_dialog_labels[0])
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torch.nn.utils.rnn import pad_sequence
batch_size = 256
class DialogDataset(torch.utils.data.Dataset):
    def __init__(self, embeddings, labels):
        self.embeddings = embeddings
        self.labels = labels

    def __len__(self):
        return len(self.embeddings)

    def __getitem__(self, idx):
        emb = torch.tensor(self.embeddings[idx], dtype=torch.float32)
        lbl = torch.tensor(self.labels[idx], dtype=torch.long)

        # Alterna speaker: 0,1,0,1,...
        spk = torch.tensor([i % 2 for i in range(len(lbl))], dtype=torch.long)

        return emb, lbl, spk

from torch.nn.utils.rnn import pad_sequence

def collate_fn(batch):
    emb_batch, lbl_batch, spk_batch = zip(*batch)

    emb_padded = pad_sequence(emb_batch, batch_first=True, padding_value=0.0)
    lbl_padded = pad_sequence(lbl_batch, batch_first=True, padding_value=-100)
    spk_padded = pad_sequence(spk_batch, batch_first=True, padding_value=0)

    lengths = torch.tensor([len(seq) for seq in lbl_batch], dtype=torch.long)

    return emb_padded, lbl_padded, spk_padded, lengths

train_dataset = DialogDataset(train_dialog_embeddings, train_dialog_labels)
val_dataset = DialogDataset(val_dialog_embeddings, val_dialog_labels)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, collate_fn=collate_fn)

test_dataset = DialogDataset(test_dialog_embeddings, test_dialog_labels)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, collate_fn=collate_fn)
```

```python
import torch
import torch.nn as nn

class TransformerClassifier(nn.Module):
    def __init__(self, input_dim=768, speaker_dim=32, model_dim=256, num_heads=2, num_layers=2,
                 num_classes=7, dropout=0.2, max_len=100):
        super(TransformerClassifier, self).__init__()

        self.speaker_embedding = nn.Embedding(2, speaker_dim)
        self.input_projection = nn.Linear(input_dim + speaker_dim, model_dim)

        self.positional_embedding = nn.Embedding(max_len, model_dim)

        encoder_layer = nn.TransformerEncoderLayer(
            d_model=model_dim,
            nhead=num_heads,
            dim_feedforward=768,
            dropout=dropout,
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        self.layer_norm = nn.LayerNorm(model_dim)
        self.dropout = nn.Dropout(dropout)

        # Nuovo classificatore MLP a 2 layer
        self.classifier = nn.Sequential(
            nn.Linear(model_dim, model_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(model_dim, num_classes)
        )

    def forward(self, x, lengths, speaker_ids):
        batch_size, seq_len, _ = x.shape

        spk_emb = self.speaker_embedding(speaker_ids)
        x_augmented = torch.cat([x, spk_emb], dim=-1)

        x_proj = self.input_projection(x_augmented)
        positions = torch.arange(seq_len, device=x.device).unsqueeze(0).expand(batch_size, seq_len)
        x_proj = x_proj + self.positional_embedding(positions)

        pad_mask = torch.arange(seq_len, device=lengths.device).expand(batch_size, seq_len) >= lengths.unsqueeze(1)
        out = self.transformer_encoder(x_proj, src_key_padding_mask=pad_mask)

        out = self.layer_norm(out)
        out = self.dropout(out)
        logits = self.classifier(out)  # MLP finale

        return logits
```

```python
model = TransformerClassifier()

```

```python
def count_parameters(model):
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Total parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")

count_parameters(model)
```

```python
import torch
from torch.optim import Adam
from sklearn.metrics import f1_score

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
criterion = nn.CrossEntropyLoss(ignore_index=-100)

model = TransformerClassifier().to(device)
optimizer = Adam(model.parameters(), lr=5e-4)

# Funzione di training
def train_epoch(model, dataloader, criterion, optimizer, scheduler, total_step, warmup_steps):
    model.train()
    total_loss = 0

    for emb_batch, labels_batch, spk_batch, lengths in dataloader:
        emb_batch = emb_batch.to(device)
        labels_batch = labels_batch.to(device)
        spk_batch = spk_batch.to(device)
        lengths = lengths.to(device)

        logits = model(emb_batch, lengths, spk_batch)
        loss = criterion(logits.view(-1, logits.size(-1)), labels_batch.view(-1))

        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        if total_step < warmup_steps:
            scheduler.step()

        total_step += 1
        total_loss += loss.item()

    return total_loss / len(dataloader), total_step



def evaluate(model, dataloader, criterion):
    model.eval()
    total_loss = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for emb_batch, lbl_batch, spk_batch, lengths in dataloader:
            emb_batch = emb_batch.to(device)
            lbl_batch = lbl_batch.to(device)
            spk_batch = spk_batch.to(device)
            lengths = lengths.to(device)

            logits = model(emb_batch, lengths, spk_batch)
            loss = criterion(logits.view(-1, logits.size(-1)), lbl_batch.view(-1))
            total_loss += loss.item()

            preds = torch.argmax(logits, dim=-1)

            mask = lbl_batch != -100
            all_preds.extend(preds[mask].cpu().numpy())
            all_labels.extend(lbl_batch[mask].cpu().numpy())


    f1_all = f1_score(all_labels, all_preds, average='macro')

    f1_wo_neutral = f1_score(all_labels, all_preds, labels=[1,2,3,4,5,6], average='macro')
    f1_micro_wo_neutral = f1_score(all_labels, all_preds, labels=[1,2,3,4,5,6], average='micro')

    acc = (np.array(all_preds) == np.array(all_labels)).mean()

    return total_loss / len(dataloader), f1_all, f1_wo_neutral, f1_micro_wo_neutral, acc
```

```python
from torch.optim.lr_scheduler import ReduceLROnPlateau
# Warmup scheduler
epochs = 80
total_step = 0
warmup_steps = 200
best_val_f1 = 0.0
patience = 16
patience_counter = 0

warmup_scheduler = torch.optim.lr_scheduler.LambdaLR(
    optimizer,
    lr_lambda=lambda step: min(1.0, step / warmup_steps)
)

# Plateau scheduler (dopo warmup)
plateau_scheduler = ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=6)


checkpoint_path = "best_model.pt"
```

```python
for epoch in range(1, epochs + 1):
    train_loss, total_step = train_epoch(model, train_loader, criterion, optimizer, warmup_scheduler, total_step, warmup_steps)
    val_loss, val_f1_all, val_f1_wo_neutral, val_f1_micro_wo_neutral,val_acc = evaluate(model, val_loader, criterion)

    print(f"Epoch {epoch}: Train Loss {train_loss:.4f} | Val Loss {val_loss:.4f} | "
          f"Acc {val_acc:.4f} | F1 (all) {val_f1_all:.4f} | F1 (wo neutral) {val_f1_wo_neutral:.4f} | F1 micro (wo neutral) {val_f1_micro_wo_neutral:.4f}")

    if total_step >= warmup_steps:
        plateau_scheduler.step(val_f1_wo_neutral)

    current_lr = optimizer.param_groups[0]['lr']
    print(f"Epoch {epoch} | Current LR: {current_lr:.6f}")

    if total_step >= warmup_steps:
        if val_f1_wo_neutral > best_val_f1:
            best_val_f1 = val_f1_wo_neutral
            patience_counter = 0
            torch.save(model.state_dict(), checkpoint_path)
            print(f"New best model saved! F1 (wo neutral): {best_val_f1:.4f}")
        else:
            patience_counter += 1
            print(f"No improvement. Patience: {patience_counter}/{patience}")

        if patience_counter >= patience:
            print("Early stopping triggered.")
            break
    else:
        print(f"Warmup phase — skipping early stopping & checkpointing. Total step: {total_step}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch

def evaluate_and_report(model, dataloader):
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for emb_batch, labels_batch, spk_batch, lengths in dataloader:
            emb_batch = emb_batch.to(device)
            labels_batch = labels_batch.to(device)
            spk_batch = spk_batch.to(device)
            lengths = lengths.to(device)

            logits = model(emb_batch, lengths, spk_batch)
            preds = torch.argmax(logits, dim=-1)
            mask = labels_batch != -100

            all_preds.extend(preds[mask].cpu().numpy())
            all_labels.extend(labels_batch[mask].cpu().numpy())

    report_dict = classification_report(
        all_labels, all_preds,
        target_names=["neutral", "anger", "disgust", "fear", "happiness", "sadness", "surprise"],
        output_dict=True
    )

    for label in report_dict:
        if isinstance(report_dict[label], dict):
            for metric in report_dict[label]:
                report_dict[label][metric] = round(report_dict[label][metric] * 100, 2)

    df_report = pd.DataFrame(report_dict).transpose()
    print("Classification Report (in percentuali):\n")
    print(df_report)

    # Confusion Matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["neutral", "anger", "disgust", "fear", "happiness", "sadness", "surprise"],
                yticklabels=["neutral", "anger", "disgust", "fear", "happiness", "sadness", "surprise"])
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix (Test Set)")
    plt.show()

evaluate_and_report(model, test_loader)
```

```python
# Normalized Confusion Matrix (per row, in percentuale)
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def evaluate_and_report_normalized(model, dataloader):
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for emb_batch, labels_batch, spk_batch, lengths in dataloader:
            emb_batch = emb_batch.to(device)
            labels_batch = labels_batch.to(device)
            spk_batch = spk_batch.to(device)
            lengths = lengths.to(device)

            logits = model(emb_batch, lengths, spk_batch)
            preds = torch.argmax(logits, dim=-1)
            mask = labels_batch != -100

            all_preds.extend(preds[mask].cpu().numpy())
            all_labels.extend(labels_batch[mask].cpu().numpy())

    # Normalized Confusion Matrix (row-wise, in percentuale)
    cm = confusion_matrix(all_labels, all_preds)
    cm_normalized = cm.astype('float') / cm.sum(axis=1, keepdims=True) * 100
    cm_normalized = np.round(cm_normalized, 2)

    plt.figure(figsize=(10, 8))  # Aumenta un po' la dimensione della figura
    sns.heatmap(
        cm_normalized,
        annot=True,
        fmt=".2f",
        cmap="Blues",
        annot_kws={"size": 14},  # Grandezza numeri nella mappa
        xticklabels=["neutral", "anger", "disgust", "fear", "happiness", "sadness", "surprise"],
        yticklabels=["neutral", "anger", "disgust", "fear", "happiness", "sadness", "surprise"]
    )
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.xlabel("Predicted", fontsize=14)
    plt.ylabel("True", fontsize=14)
    plt.title("Normalized Confusion Matrix (Test Set) - Percentuali", fontsize=16)
    plt.tight_layout()  # Migliora la disposizione
    plt.show()

evaluate_and_report_normalized(model, test_loader)
```
