```python
import osmnx as ox
import geopandas as gpd
import gdown

# Download and load the 250 m grid (reprojected to EPSG:4326 for OSM compatibility)
gdown.download("https://drive.google.com/uc?id=1GMLbyT-iN6cqubGLOJX2rexzTmLNSbTD", "grid_250m_with_nil_clean.geojson", quiet=False)
grid = gpd.read_file("grid_250m_with_nil_clean.geojson").to_crs(epsg=4326)
```

```python
import osmnx as ox

# Download pedestrian network for Milan using official administrative boundary
G = ox.graph_from_place("Milano, Lombardia, Italia", network_type="walk", simplify=True)

# Save graph to file
ox.save_graphml(G, "milano_walk.graphml")
```

```python
ox.plot_graph(G, bgcolor="white", node_size=5, edge_color="#999999")
```
